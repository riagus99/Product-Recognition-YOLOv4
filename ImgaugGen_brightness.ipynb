# modified, modified_color 폴더에 있는 사진들 밝기 변화

```python
import os, random
import numpy as np
import cv2
import imgaug as ia
from imgaug import augmenters as iaa
import imageio

ia.seed(1)
```

```python
base_dir = '/home/ai_competition36'
```

```python
import shutil
```

```python
for i in range(4,5):
    target_dir = os.path.join(base_dir, f'1.competition_trainset/modified{i}')
    img_list = [file_name for file_name in os.listdir(target_dir) if file_name.endswith(".jpg")]
    save_dir = os.path.join(base_dir, f'1.competition_trainset/modified{i}_brightness')
    N = len(img_list)
    height = 480
    width = 640
    channels = 3
    images = np.zeros((N, height, width, channels), dtype=np.uint8)
    
    for j, img in enumerate(img_list):                                  # 이미지 데이터를 미리 읽어들여 images 4차원 배열에 저장 
        img_path = os.path.join(target_dir, img)
        images[j, :, :, :] = imageio.imread(img_path)
        if j % 1000 == 0:
            print(f'{j}times iterating...')
            
    print(f'modified{i} img loading complete')               
    
    from imgaug import augmenters as iaa
    brightness=iaa.AddToBrightness((-50,75))
    images_aug_brightness=brightness(images=images)
    
    for k in range(N):                                                  # 밝기 변화 학습데이터
        imageio.imwrite(save_dir + f'/{k}+background_brightness.jpg', images_aug_brightness[k])
        img_file_name = img_list[k]
        txt_file_name = img_file_name[0:-3] + 'txt'                                # img에 대응되는 txt 파일을 찾는다.
        src = os.path.join(target_dir, txt_file_name)
        dst = os.path.join(save_dir, f'{k}+background_brightness.txt')                     # img파일과 같은 이름으로 txt파일 복사해서 저장
        shutil.copy(src, dst)
        if k % 1000 == 0:
            print('{}times augmenting files...'.format(k))
            
    print(f'modified{i} complete')
    
print('complete')
```

```python
for i in range(4,5):
    target_dir = os.path.join(base_dir, f'1.competition_trainset/modified{i}_color')
    img_list = [file_name for file_name in os.listdir(target_dir) if file_name.endswith(".jpg")]
    save_dir = os.path.join(base_dir, f'1.competition_trainset/modified{i}_brightness')
    N = len(img_list)
    height = 480
    width = 640
    channels = 3
    images = np.zeros((N, height, width, channels), dtype=np.uint8)
    
    for j, img in enumerate(img_list):                                  # 이미지 데이터를 미리 읽어들여 images 4차원 배열에 저장 
        img_path = os.path.join(target_dir, img)
        images[j, :, :, :] = imageio.imread(img_path)
        if j % 1000 == 0:
            print(f'{j}times iterating...')
            
    print(f'modified{i}_color img loading complete')               
    
    
    brightness=iaa.AddToBrightness((-50,75))
    images_aug_brightness=brightness(images=images)
    
    for k in range(N):                                                  # 밝기 변화 학습데이터
        imageio.imwrite(save_dir + f'/{k}+background_hue_brightness.jpg', images_aug_brightness[k])
        img_file_name = img_list[k]
        txt_file_name = img_file_name[0:-3] + 'txt'                                # img에 대응되는 txt 파일을 찾는다.
        src = os.path.join(target_dir, txt_file_name)
        dst = os.path.join(save_dir, f'{k}+background_hue_brightness.txt')                     # img파일과 같은 이름으로 txt파일 복사해서 저장
        shutil.copy(src, dst)
        if k % 1000 == 0:
            print('{}times augmenting files...'.format(k))
            
    print(f'modified{i}_color complete')
    
print('complete')
```
